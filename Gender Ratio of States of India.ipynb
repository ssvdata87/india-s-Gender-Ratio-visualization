```python
from chart_studio import plotly
import plotly 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('gender_ratio_india.csv')
data.head() #GR=Gender Ratio
```

```python
import matplotlib.pyplot as plt
import seaborn
df=data['GR']
STATE=data['STATE']
fig = plt.figure(figsize =(20, 10))
plt.pie(df, labels = STATE)
 
# show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 


fig, ax = plt.subplots(figsize=(15, 6))


sns.barplot(x='STATE',y='GR',data = data, order=data.sort_values('GR',ascending = False).STATE,ax=ax)

 
plt.xticks(rotation=90)


plt.show()
```
